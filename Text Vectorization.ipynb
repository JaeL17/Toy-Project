```python
import pandas as pd
```

```python
#import datasets
df_train = pd.read_csv('movie_review_train.csv')
df_test = pd.read_csv('movie_review_test.csv')
```

```python
df_train
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(df_train['Review'].values)
y_train = df_train['Label']
```

```python
pd.DataFrame(X_train.toarray(), columns = vectorizer.get_feature_names())
```

```python
X_test=vectorizer.transform(df_test['Review'].values)
y_test=df_test['Label']
```

```python
# Saving the results
import scipy.sparse
scipy.sparse.save_npz('X_train_sparse_matrix.npz', X_train)
scipy.sparse.save_npz('X_test_sparse_matrix.npz', X_test)
```
